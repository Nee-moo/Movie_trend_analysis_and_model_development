```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import ast
```

```python
# chuyển str -> list
df['Genres'] = df['Genres'].apply(ast.literal_eval)
```

```python
genre_counts = df['Genres'].explode().value_counts()
genre_counts
```

```python
total = genre_counts.sum()
total
```

```python
# phân phối thể loại

min_percent = 0.01  # xuất hiện dưới 1% = Other

rare_genres = genre_counts[genre_counts / total < min_percent]
popular_genres = genre_counts[genre_counts / total >= min_percent]

genre_counts_final = popular_genres.copy()
genre_counts_final['Other'] = rare_genres.sum()

genre_counts_final = genre_counts_final.sort_values(ascending=False) # Sắp xếp giảm dần
genre_counts_final
```

```python
plt.figure(figsize=(8, 8))

plt.pie(
    genre_counts_final,
    labels=genre_counts_final.index,
    autopct='%1.1f%%',
    startangle=140,
)
plt.title('Phân bổ thể loại')
plt.show()
```

```python
# Heatmap
numeric_columns = ["$Worldwide", "$Domestic", "Domestic %", "$Foreign", "Foreign %", "Rating", "Vote_Count"]
df_numeric = df[numeric_columns]
f, ax = plt.subplots(figsize=(10, 6))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5, ax=ax)
plt.show()
```

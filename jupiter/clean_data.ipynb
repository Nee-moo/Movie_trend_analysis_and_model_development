```python
import pandas as pd 
import numpy as np 
import re
```

```python
# Load data
df = pd.read_excel('../data/raw/Movie_dataset.xlsx')
df.head()  
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.shape
```

```python
# Đổi tên và sắp xếp 
df.rename(columns={"Release Group":"Title"}, inplace = True)
df=df[["Title","Genres","Year","Rating","Vote_Count","Production_Countries","$Worldwide","$Domestic","Domestic %","$Foreign","Foreign %"]]
df.head()
```

```python
df.dtypes
```

```python
# Chuyển về số nguyên
df['Year']= df['Year'].astype(int)
df['Vote_Count']= df['Vote_Count'].astype(int)
df['$Worldwide']= df['$Worldwide'].astype(int)
df['$Domestic']= df['$Domestic'].astype(int)
df['$Foreign']= df['$Foreign'].astype(int)
```

```python
# Tách chuỗi và bỏ khoảng trắng
df['Genres'] = df['Genres'].str.split(',')
df['Genres'] = df['Genres'].apply(lambda x: [item.strip() for item in x])
df['Genres'] = df['Genres'].apply(lambda x: list(set(x))) # Bỏ trùng lặp

df['Production_Countries'] = df['Production_Countries'].str.split(',')
df['Production_Countries'] = df['Production_Countries'].apply(lambda x: [item.strip() for item in x])
df['Production_Countries'] = df['Production_Countries'].apply(lambda x: list(set(x)))
```

```python
# Bỏ /10 và lấy 1 số sau dấu phẩy
df["Rating"]= (df["Rating"].str.replace("/10","",regex=False).astype(float).round(1))
df.head()
```

```python
df['Domestic %']= df['Domestic %'].round(1)
df['Foreign %']= df['Foreign %'].round(1)
df.head()
```

```python
df.to_csv('cleaned_data.csv',index=False)
```
